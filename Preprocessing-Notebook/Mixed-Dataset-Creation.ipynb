```python
import os
import random
import shutil
from tqdm import tqdm
```

#### Making test dataset
#### rest will be used for training

```python
def copy_files(source, destination, total, prefix='d_'):
    files= os.listdir(source)
    selected_files= random.sample(files, total)
    i=1
    for file in selected_files:
        s_file= os.path.join(source, file)
        rename_file=prefix+file
        d_file= os.path.join(destination, rename_file)

        shutil.copy2(s_file, d_file)
        print(f"{i}  : Copied {file} to {d_file}")
        i+=1
    
```

```python
source='Datasets/DSDD/images/'
destination='Datasets/Mixed/images/'
copy_files(source, destination, 567, 'd_')
```

```python
source='Datasets/SSDD/images/'
destination='Datasets/Mixed/images/'
copy_files(source, destination, 567, 's_')
```

```python
def copying_txt_file(dsdd_source,ssdd_source, destination, mixed_file_loc):
    files= os.listdir(mixed_file_loc)
    dsdd= [file for file in files if file.startswith('d_')]
    ssdd= [file for file in files if file.startswith('s_')]
    for file in tqdm(dsdd):
        file_name= file.split('_')[-1].split('.')[0]+'.txt'
        s_file= os.path.join(dsdd_source, file_name)
        rename_file='d_'+file_name
        d_file= os.path.join(destination, rename_file)
        shutil.copy2(s_file, d_file)
    for file in tqdm(ssdd):
        file_name= file.split('_')[-1].split('.')[0]+'.txt'
        s_file= os.path.join(ssdd_source, file_name)
        rename_file='s_'+file_name
        d_file= os.path.join(destination, rename_file)
        shutil.copy2(s_file, d_file)
    print('Copying Finish')
        
```

```python
dsdd_source='Datasets/DSDD/labels/'
ssdd_soruce= 'Datasets/SSDD/labels/'
destination='Datasets/Mixed/labels/'
mixed_file= 'Datasets/Mixed/images/'
copying_txt_file(dsdd_source,ssdd_soruce,destination,mixed_file)
```

### Training set

```python
def training_images(source, destination, filtered_files, prefix='d_'):
    files= os.listdir(filtered_files)
    dsdd= [file for file in files if file.startswith(prefix)]
    dsdd= [file.split('_')[-1] for file in dsdd]
    orginal_list= os.listdir(source)
    set1= set(dsdd)
    set2= set(orginal_list)
    remain_files= set2-set1
    for file in tqdm(remain_files):
        s_file= os.path.join(source, file)
        rename_file=prefix+file
        d_file= os.path.join(destination, rename_file)
        shutil.copy2(s_file, d_file)
        
    
```

```python
source='Datasets/DSDD/images/'
destination='Datasets/Mixed/Train/images/'
current_folder='Datasets/Mixed/Test/images/'
training_images(source, destination, current_folder, 'd_')
```

```python
source='Datasets/SSDD/images/'
destination='Datasets/Mixed/Train/images/'
current_folder='Datasets/Mixed/Test/images/'
training_images(source, destination, current_folder, 's_')
```

```python
source='Datasets/DSDD/labels/'
destination='Datasets/Mixed/Train/labels/'
current_folder='Datasets/Mixed/Test/labels/'
training_images(source, destination, current_folder, 'd_')
################################################################
source='Datasets/SSDD/labels/'
destination='Datasets/Mixed/Train/labels/'
current_folder='Datasets/Mixed/Test/labels/'
training_images(source, destination, current_folder, 's_')
```
